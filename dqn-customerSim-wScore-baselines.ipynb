# Compare to Baselines
### 1) Historical Baseline

```python
import pandas as pd
import numpy as np

data = pd.read_csv('C:\\Users\\renaza\\OneDrive - SAS\\kdd1998\\kdd1998tuples.csv',
               header=None)

data.columns = ["customer", "period", "_S0_", "_S1_", "_S2_", "_S3_", "_S4_", "_S5_", "_S6_", "_S7_",
                "_S8_", "zip_la", "zip_lo", "a", "rew", "r1", "f1", "m1", "ir1", "if1", "gender", "age",
                "income", "zip_region", "zip_la", "zip_lo"]
```

```python
customer_idx = set(data['customer'])
reward_lst_hist = []
for i in range(nBaselineEP):
    data_cust = data[data["customer"] == np.random.randint(len(customer_idx))]
    reward_lst_hist.append(list(data_cust['rew'])[:18])
```

```python
baseline_mean_rew_hist = np.mean(np.sum(reward_lst_hist,1))
print("Mean cumulative historical reward is : ", baseline_mean_rew_hist)
```

### 2) Random policy

```python
import numpy as np
import gym 
import gym_customersim
```

```python
env = gym.make('CustomerSim-v0')
```

```python
reward_lst_rand = []
for i in range(nBaselineEP):
    rew_ep = []
    state = env.reset()
    done = False
    while not done:
        a = np.random.randint(12)
        nstate, r, done, _ = env.step(a)
        rew_ep.append(r)
    reward_lst_rand.append(rew_ep)
    
```

```python
baseline_mean_rew_rand = np.mean(np.sum(reward_lst_rand,1))
print("Mean cumulative reward is : ", baseline_mean_rew_rand)
```

### Summary 

```python
print("RL cumulative reward is : ", rl_mean_rew)
print("Mean cumulative historical reward is : ", baseline_mean_rew_hist)
print("Mean cumulative reward is : ", baseline_mean_rew_rand)

print('============')
print("Ratio of RL over historical policy : ", rl_mean_rew/baseline_mean_rew_hist)
print("Ratio of RL over random policy : ", rl_mean_rew/baseline_mean_rew_rand)
```

```python
# average reward at each time
rl_rew_t = np.reshape(np.array(list(rl_rew)),[18,-1])
rl_rew_t = np.sum(rl_rew_t, 1)/nBaselineEP

reward_lst_hist_t = np.sum(reward_lst_hist,0)/nBaselineEP
reward_lst_rand_t = np.sum(reward_lst_rand,0)/nBaselineEP

# cumulative reward over time
rl_rew_ct = np.cumsum(rl_rew_t)
reward_lst_hist_ct = np.cumsum(reward_lst_hist_t)
reward_lst_rand_ct = np.cumsum(reward_lst_rand_t)
```

```python
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "serif"
plt.rcParams["figure.figsize"] = (8,4)

time_range = np.arange(1,19)
plt.plot(time_range, rl_rew_ct, label="RL")
plt.plot(time_range, reward_lst_hist_ct, label="Historical")
plt.plot(time_range, reward_lst_rand_ct, label="Random")

plt.ylabel("Average Reward", fontsize=12)
plt.xlabel("Time", fontsize=12)
plt.title("Cumulative average reward over time", fontsize=14)

plt.axes().xaxis.set_major_locator(MaxNLocator(integer=True))

plt.legend()
plt.show()
```
